```python
train_ = train_set
test_ = test_set
actual = []
predict = []
STEPS = 20
for i in range(STEPS):
    actual.append(test_[i,0])
    print(train_[-1])
    train_ = np.vstack((train_,test_[i]))
    print(train_[-1])
    print(len(train_))
    print(actual)
    train_X,train_y = util.get_reshaped_data_by_OP(offset=50,period=22,data=train_,is_scaled=True)
    print(train_y[-1])
    y_hat_ = []
    for j in range(1):  
        model = estm_xgb
        X_,_,y_,_ = train_test_split(train_X[:-1],train_y[:-1],test_size=0.5,random_state=i*j+j)
        #model.fit(train_X[:-1],train_y[:-1])
        model.fit(X_,y_)
        y_hat_.append(model.predict(train_X[-1].reshape(1,-1))[0])
    y_pre = np.mean(y_hat_)
    predict.append(y_pre)
    print(f'actual: {actual[i]}, predict: {y_pre}')
    train_[-1,0] = y_pre
```
